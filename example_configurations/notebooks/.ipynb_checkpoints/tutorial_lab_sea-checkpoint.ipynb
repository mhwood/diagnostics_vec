# Tutorial: Lab Sea

```python
import numpy as np
from MITgcmutils import mds
import matplotlib.pyplot as plt
import netCDF4 as nc4

directory = '../tutorial_lab_sea'
run_dir = directory+'/run'

n_depths = 23
n_rows = 16
n_cols = 20
n_ob_timesteps = 9

# mask dimensions
min_col = 8
max_col = 11
min_row = 7
max_row = 13
```

#### The Output of lab_sea is via mnc - this function is defined to combine variables from the separate files to a single field

```python
def combine_variables_across_mnc_outputs(run_dir,file_prefix,variable_name,variable_shape):
    
    # read in the first variable
    file_path_1 = run_dir+'/mnc_test_0001/'+file_prefix+'.t002.nc'
    ds_1 = nc4.Dataset(file_path_1)
    if variable_shape=='4D':
        var_1 = ds_1.variables[variable_name][:,:,:,:]
    if variable_shape=='3D':
        var_1 = ds_1.variables[variable_name][:,:,:]
    if variable_shape=='2D':
        var_1 = ds_1.variables[variable_name][:,:]
    
    # read in the second variable
    file_path_2 = run_dir+'/mnc_test_0002/'+file_prefix+'.t001.nc'
    ds_2 = nc4.Dataset(file_path_2)
    if variable_shape=='4D':
        var_2 = ds_2.variables[variable_name][:,:,:,:]
    if variable_shape=='3D':
        var_2 = ds_2.variables[variable_name][:,:,:]
    if variable_shape=='2D':
        var_2 = ds_2.variables[variable_name][:,:]
    
    var = np.concatenate([var_2,var_1],axis=-1)
    return(var)
```

#### Use the function defined above to read in X and Y for the plots below

```python
x = combine_variables_across_mnc_outputs(run_dir,'grid','XC','2D')
y = combine_variables_across_mnc_outputs(run_dir,'grid','YC','2D')
```

## Visualize the Domain and Masks

```python
bathymetry = np.fromfile(directory+'/input/bathy.labsea1979', dtype='>f4')
print(np.shape(bathymetry))
bathymetry = np.reshape(bathymetry,(n_rows,n_cols))

fig = plt.figure(figsize=(12,8))

plt.subplot(2,1,1)
C = plt.contourf(x,y,bathymetry,50,cmap='Blues_r')
C_tmp = plt.contour(x,y,bathymetry,levels=[-10],colors='k')
cbar = plt.colorbar(C)
cbar.set_label('Depth (m)', rotation=90)
plt.title('Bathymetry')

plt.subplot(2,2,3)
all_boundary_masks = np.zeros_like(bathymetry)
for mask_name in ['lateral_mask_east','lateral_mask_west','lateral_mask_north','lateral_mask_south']:
    mask = np.fromfile(directory+'/input/'+mask_name+'.bin', dtype='>f4')
    mask = np.reshape(mask,(n_rows,n_cols))
    all_boundary_masks[mask>0]=1
#all_boundary_masks = np.ma.masked_where(mask == 0, mask)
plt.imshow(all_boundary_masks,extent=[np.min(x),np.max(x),np.min(y),np.max(y)],origin='lower')
C_tmp = plt.contour(x,y,bathymetry,levels=[-10],colors='yellow')
plt.title('Boundary Masks')


plt.subplot(2,2,4)
mask_name = 'surface_mask'
mask = np.fromfile(directory+'/input/'+mask_name+'.bin', dtype='>f4')
mask = np.reshape(mask,(n_rows,n_cols))
mask[mask>0]=1
#mask = np.ma.masked_where(mask == 0, mask)
plt.imshow(mask,extent=[np.min(x),np.max(x),np.min(y),np.max(y)],origin='lower')
C_tmp = plt.contour(x,y,bathymetry,levels=[-10],colors='yellow')
plt.title('Surface Mask')
```

## Compare sea ice variables on surface

```python
time_step = 10

fig=plt.figure(figsize=(15,20))

si_vars = ['AREA','HEFF','UICE','VICE']

counter=1
for var in si_vars:

    diagnostics_var = combine_variables_across_mnc_outputs(run_dir,'sice.0000000001','si_'+var,'3D')

    diagnostics_ob_var = np.fromfile(directory+'/run/MASK_SF_01_'+var+'.bin',dtype='>f4')
    diagnostics_ob_var = np.reshape(diagnostics_ob_var,(n_ob_timesteps,max_row-min_row+1,max_col-min_col+1))


    # plot southern boundary first
    plt.subplot(len(si_vars),3,counter)
    C = plt.contourf(x[min_row:max_row+1,min_col:max_col+1],y[min_row:max_row+1,min_col:max_col+1],
                     diagnostics_var[-1,min_row:max_row+1,min_col:max_col+1],50)
    cbar = plt.colorbar(C)
    cbar.set_label(var)
    plt.ylabel('Latitude (deg)')
    if counter==(len(si_vars)-1)*3+1:
        plt.xlabel('Longitude (deg)')
    if counter==1:
        plt.title('Diagnostics Output')
    counter+=1

    # plot the diagnostics_ob output second
    plt.subplot(len(si_vars),3,counter)
    C2 = plt.contourf(x[min_row:max_row+1,min_col:max_col+1],y[min_row:max_row+1,min_col:max_col+1],
                      diagnostics_ob_var[-1,:,:],50)
    cbar = plt.colorbar(C2)
    cbar.set_label(var)
    if counter==(len(si_vars)-1)*3+2:
        plt.xlabel('Longitude (deg)')
    if counter==2:
        plt.title('Diagnostics_ob Output')
    counter+=1  

    # plot the difference
    plt.subplot(len(si_vars),3,counter)
    C3 = plt.contourf(x[min_row:max_row+1,min_col:max_col+1],y[min_row:max_row+1,min_col:max_col+1],
                      diagnostics_var[-1,min_row:max_row+1,min_col:max_col+1]-diagnostics_ob_var[-1,:,:],50,cmap='seismic')
    cbar = plt.colorbar(C3)
    cbar.set_label(var)
    if counter==(len(si_vars)-1)*3+3:
        plt.xlabel('Longitude (deg)')
    if counter==3:
        plt.title('Difference')
    counter+=1


    plt.suptitle('Standard Output vs. Diagnostics_ob Output Comparison')
```
