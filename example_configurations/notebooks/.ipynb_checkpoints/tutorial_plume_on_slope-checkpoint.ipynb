# Tutorial: Plume on a Slope

```python

import os
folder='../tutorial_plume_on_slope'


print(os.listdir(folder))
```

```python
import numpy as np
from MITgcmutils import mds
import matplotlib.pyplot as plt
directory = '../tutorial_plume_on_slope'

x = mds.rdmds(directory+'/run/XC*').ravel()
z = mds.rdmds(directory+'/run/RC')[:,0,0]
print(np.shape(x))
print(np.shape(z))
```

```python
time_step = 20

x_subset = x[:161]
z_subset = z[:15]

standard_T = mds.rdmds(directory+'/run/T*',time_step)
standard_T = standard_T[:,0,:]
standard_T_subset = standard_T[:15,:161]

diagnostics_ob_T = np.fromfile(directory+'/run/MASK_OB_01_THETA.bin',dtype='>f8')
diagnostics_ob_T = np.reshape(diagnostics_ob_T,(20,15,161))

fig=plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
C = plt.contourf(x_subset,z_subset,standard_T_subset,50)
plt.colorbar(C)
plt.ylabel('Depth (m)')
plt.xlabel('Distance Along Profile (m)')
plt.title('Standard Output')

plt.subplot(1,3,2)
C2 = plt.contourf(x_subset,z_subset,diagnostics_ob_T[-1,:,:],50)
plt.colorbar(C2)
plt.xlabel('Distance Along Profile (m)')
plt.title('Diagnostics_ob Output')

plt.subplot(1,3,3)
C3 = plt.contourf(x_subset,z_subset,standard_T_subset-diagnostics_ob_T[-1,:,:],50,cmap='seismic')
plt.colorbar(C3)
plt.xlabel('Distance Along Profile (m)')
plt.title('Difference')


plt.suptitle('Standard Output vs. Diagnostics_ob Output Comparison')
```
